```python
df2 = df[df['fwd_id'].isna()].groupby(['peer_id'])['peer_id'].count()
most_message_channels=df2[df2 > 300]
```

```python
import re

regex = r"Ø¯Ù„Ø§Ø±.+ØªÙ‡Ø±Ø§Ù† (\d+)"
```

```python
output_df=pd.DataFrame([],columns=['peer_username','date','price'])
for index,row in nlp_df.iterrows():
    test_str=row['text']
    x = re.search(regex, str(test_str))
    if x:
        price =x.groups()[0]
        if len(price) > 4:
            new_row={
                'peer_username':row['peer_username'],
                'date':row['date'],
                'price':price
            }
            output_df=pd.concat([output_df, pd.DataFrame([new_row])], ignore_index=True)
print('done !')
```
